```python
#Data cleaning and Feature Engineering 
df.isna().sum()
```

```python
df['Day_no'] = df['Date_of_Journey'].dt.dayofweek
df = df.drop(labels=['Date_of_Journey','Day'],axis=1)
```

```python
# Arrival Feature Engineering
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda s:s.split(' ')[0])
df['Arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min'] = df['Arrival_Time'].str.split(':').str[1]
df['Arrival_hour'] = df['Arrival_hour'].astype(int)
df['Arrival_min'] = df['Arrival_min'].astype(int)
df = df.drop(labels=['Arrival_Time'],axis=1)
df.head()
```

```python
# Departure Feature Engineering
df['Dep_hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dep_min'] = df['Dep_Time'].str.split(':').str[1]
df['Dep_hour'] = df['Dep_hour'].astype(int)
df['Dep_min'] = df['Dep_min'].astype(int)
df = df.drop(labels=['Dep_Time'],axis=1)
```

```python
# Creating Function to get the duration of flight from string 
def get_duration_float(s):
    if ((s.find('h')>0) and (s.find('m')>0)):
        hrs = s.split('h')[0]
        mins = s.split(' ')[1][:-1]
    elif((s.find('h')>0) and (s.find('m')<0)):
        hrs = s.split('h')[0]
        mins = 0
    elif((s.find('h')<0) and (s.find('m')>0)):
        hrs = 0
        mins = s.split('m')[0]
    duration = float(hrs) + float(mins)/60
    return duration
df['dur_float'] = df['Duration'].apply(get_duration_float)
df = df.drop(labels=['Duration'],axis=1)
df.head(2)
```

```python
# Total Stops
df[df['Total_Stops'].isna()]
```

```python
md = df['Total_Stops'].mode()[0]
md
```

```python
df['Total_Stops'] = df['Total_Stops'].fillna(md)
```

```python
df['Total_Stops'].isna().sum()
```

```python
stops_mapper = {
    'non-stop':0,
    '1 stop':1,
    '2 stops':2,
    '3 stops':3,
    '4 stops':4
}
df['Total_Stops'] = df['Total_Stops'].map(stops_mapper)
df.head()
```

```python
# Drop route and additional info
df = df.drop(labels=['Route','Additional_Info'],axis=1)
df.head(2)
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates(keep='first')
```

```python
# Save Clean File as csv
df.to_csv('flight_cleaned.csv')
```

### Feature Engineering 

```python
# Seperating X and Y
X = df.drop(labels=['Price'],axis=1)
Y = df[['Price']]
```

```python
# Seperating Numeric and Categorical Features
cat_cols = list(X.select_dtypes(include='object').columns)
num_cols = list(X.select_dtypes(exclude='object').columns)
print(f'Numerical Features : {num_cols}')
print(f'Categorical Features : {cat_cols}')
```

```python
# One Hot Encoding of Categorical Variables
X_cat = pd.get_dummies(X[cat_cols])
X_cat.head()
```

```python
# Concatenating Num and Cat variables
X_concat = pd.concat([X[num_cols],X_cat],axis=1)
X_concat.head()
```

```python
# Train Test split
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_concat,Y,test_size=0.2,random_state=23)
```

```python
# Apply Standard Scaling on X
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
xtrain = pd.DataFrame(scaler.fit_transform(xtrain),columns=xtrain.columns)
xtest = pd.DataFrame(scaler.transform(xtest),columns=xtest.columns)
```

```python
xtrain.head()
```

```python
xtest.head()
```

```python
pip install catboost
```

```python
# Catboost Regressor 
from catboost import CatBoostRegressor
model = CatBoostRegressor(loss_function='RMSE',iterations=1000)
model.fit(xtrain,ytrain)
```

```python
# Predicting Values
ytrain_pred = model.predict(xtrain)
ytest_pred = model.predict(xtest)
```

```python
# Function to evaluate metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
def evaluate_model(actual, predicted):
    mae = mean_absolute_error(actual,predicted)
    mse = mean_squared_error(actual,predicted)
    rmse = mse**0.5
    r2 = r2_score(actual,predicted)
    return mae, rmse, r2
```

```python
# Function to print model metrics
def print_eval_results(ytrain,ytrain_pred,ytest,ytest_pred):
     # Evaluate Train and Test dataset
    model_train_mae , model_train_rmse, model_train_r2 = evaluate_model(ytrain, ytrain_pred)
    model_test_mae , model_test_rmse, model_test_r2 = evaluate_model(ytest, ytest_pred)
    # Printing results
    print('Model performance for Training set')
    print("- Root Mean Squared Error: {:.4f}".format(model_train_rmse))
    print("- Mean Absolute Error: {:.4f}".format(model_train_mae))
    print("- R2 Score: {:.4f}".format(model_train_r2))
    print('----------------------------------')
    
    print('Model performance for Test set')
    print("- Root Mean Squared Error: {:.4f}".format(model_test_rmse))
    print("- Mean Absolute Error: {:.4f}".format(model_test_mae))
    print("- R2 Score: {:.4f}".format(model_test_r2))
```

```python
print_eval_results(ytrain, ytrain_pred, ytest, ytest_pred)
```

```python
# Model Importance
imp = pd.Series(model.feature_importances_)
imp.index = model.feature_names_
imp = imp.sort_values(ascending=False)
imp
```

```python
imp.sort_values().plot(kind='barh',figsize=(10,8), title = 'Catboost Feature Importances')
```
