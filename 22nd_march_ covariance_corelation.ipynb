## Q2. Spearman's rank correlation is a measure of the monotonic relationship between two variables. Suppose you have collected data on the amount of sleep individuals get each night and their overall job satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two variables and interpret the result.

```python
# Ans:
    
from scipy.stats import spearmanr
# Define the two variables as lists
sleep = [6, 7, 8, 5, 7, 6, 8, 9, 5, 7, 6, 8, 7, 6, 5, 9, 8, 7, 5, 6]
job_satisfaction = [5, 7, 8, 4, 7, 6, 9, 10, 4, 7, 5, 8, 7, 6, 4, 9, 8, 7, 5, 6]
# Calculate the Spearman's rank correlation coefficient and p-value using scipy's spearmanr function
corr_coef, p_value = spearmanr(sleep, job_satisfaction)
print("The Spearman's rank correlation coefficient between sleep and job satisfaction is:", corr_coef)
print("The p-value for this correlation is:", p_value)
```

This indicates a positive monotonic relationship between the two variables.

## Q5. A survey was conducted to examine the relationship between age and preference for a particular brand of soft drink. The survey results are shown below:

Age(Years):
25,
42,
37,
19,
31,
28


Soft drink Preferences:
coke,
pepsi,
Coke,
Pepsi,
Coke.


```python
import pandas as pd
data = {'Age': [25, 42, 37, 19, 31, 28],
        'Soft Drink Preference': ['Coke', 'Pepsi', 'Mountain Dew', 'Coke', 'Pepsi', 'Coke']}
df = pd.DataFrame(data)
# Perform f oneway test
from scipy.stats import f_oneway
coke = df[df['Soft Drink Preference'] == 'Coke']['Age']
pepsi = df[df['Soft Drink Preference'] == 'Pepsi']['Age']
mountain_dew = df[df['Soft Drink Preference'] == 'Mountain Dew']['Age']
f_statistic, p_value = f_oneway(coke, pepsi, mountain_dew)
print('F-statistic:', f_statistic)
print('p-value:', p_value)
```

```python
# Assuming alpha value of 0.05 
alpha = 0.05
if p_value < alpha:
    print('Reject the Null Hypothesis')
    print('Conclusion : Age and Soft Drink Preference are DEPENDENT of each other')
else:
    print('FAIL to reject the Null Hypothesis')
    print('Conclusion : Age and Soft Drink Preference are INDEPENDENT of each other')
```

```python
import seaborn as sns
sns.boxplot(data=df,x='Soft Drink Preference',y='Age')
```

## Q6. A company is interested in examining the relationship between the number of sales calls made per day and the number of sales made per week. The company collected data on both variables from a sample of 30 sales representatives. Calculate the Pearson correlation coefficient between these two variables.

```python
##Ans:
import pandas as pd    
import numpy as np
from scipy.stats import pearsonr
# Define the two variables as lists
sales_calls = [20, 25, 30, 35, 40, 15, 22, 28, 33, 38, 21, 27, 32, 37, 18, 24, 29, 34, 39, 16, 23, 26, 31, 36, 19, 27, 33, 38, 22, 29]
sales_made = [2, 4, 5, 6, 7, 1, 3, 4, 5, 7, 2, 4, 5, 6, 1, 3, 4, 6, 7, 1, 3, 4, 5, 6, 2, 4, 5, 6, 3, 5]
df = pd.DataFrame({'sales_calls':sales_calls,'sales_made':sales_made})

# Calculate the Pearson correlation coefficient and p-value using scipy's pearsonr function
corr_coef, p_value = pearsonr(sales_calls, sales_made)
print("The Pearson correlation coefficient between sales calls and sales made is:", corr_coef)
print("The p-value for this correlation is:", p_value)
```

The correlation coefficient is positive and strong (0.98), indicating a strong positive relationship between the two variables (i.e., as the number of sales calls increases, the number of sales made also tends to increase).

```python
import seaborn as sns
sns.regplot(data=df, x = 'sales_calls', y='sales_made',line_kws={'color':'red'},ci=95)
```
