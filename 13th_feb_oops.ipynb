Q1. Explain why we have to use the Exception class while creating a Custom Exception.

Ans:

The Exception class is the base class for all built-in exceptions in Python. When creating a custom exception, it is recommended to inherit from the Exception class so that the custom exception can be handled just like any other built-in exception. By inheriting from the Exception class, the custom exception inherits all the methods and properties of the base class, allowing it to be customized and tailored to the specific use case.

Also, using the Exception class ensures that the custom exception is compatible with the try/except structure used to catch exceptions in Python, making it easier to handle and debug errors in code.

Ans: The LookupError exception in Python forms the base class for all exceptions that are raised when an index or a key is not found for a sequence or dictionary respectively.
So LookupError exception class can handle both IndexError and KeyError exception classes.

1.Key Error:

```python
L1 = {'name': 'Rajat kanwar','age': 28,'language': 'Python'}
user_input = input('Enter the key')
try:
    print(f'{user_input} is {L1[user_input]}')
except LookupError as e:
    print(f'{e}, {e.__class__}')
```

Here the user input key is not present in the dictionary.

2.Index error:

An IndexError means that your code is trying to access an index that is invalid.
For example, if you have a list with three items and you try to access the fourth item, you will get an IndexError.

```python
myList = [1, 2, 3, 4, 5]
print("The list is:", myList)
index = 5
element = myList[index]
print("Element at index {} is {}".format(index,element))
```
