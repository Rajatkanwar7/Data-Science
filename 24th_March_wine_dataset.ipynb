## Q1. What are the key features of the wine quality data set? Discuss the importance of each feature in predicting the quality of wine.

Ans:
    
Some of the key features of this dataset include:
1. Fixed acidity: The fixed acidity of a wine refers to the amount of non-volatile acids present in the wine. These acids contribute to the overall tartness and acidity of the wine, and can play a role in determining its overall quality.
2. Volatile acidity: Volatile acidity refers to the amount of volatile acids present in the wine. These acids can contribute to off-flavors and unpleasant aromas in the wine, and high levels of volatile acidity can indicate a lower-quality wine.
3. Citric acid: Citric acid is one of the primary acids found in wine, and can contribute to its overall acidity and freshness. Higher levels of citric acid may indicate a higher-quality wine.
4. Residual sugar: Residual sugar refers to the amount of sugar remaining in the wine after fermentation. Wines with higher levels of residual sugar may be perceived as sweeter and can have a different mouthfeel than wines with lower levels of residual sugar.
5. Chlorides: Chlorides are a measure of the salt content in the wine, and can contribute to its overall taste and mouthfeel. High levels of chlorides can indicate a lower-quality wine.
6. Free sulfur dioxide: Free sulfur dioxide is added to wines as a preservative and can help prevent spoilage. However, high levels of free sulfur dioxide can contribute to unpleasant aromas and flavors in the wine.
7. Total sulfur dioxide: Total sulfur dioxide refers to the total amount of sulfur dioxide in the wine, including both free and bound sulfur dioxide. High levels of total sulfur dioxide can indicate a lower-quality wine.
8. Density: Density is a measure of the wine's mass per unit volume, and can indicate its alcohol content or residual sugar levels.
9. pH: pH is a measure of the acidity of the wine. Wines with a lower pH may be perceived as more acidic and tart, while wines with a higher pH may be perceived as smoother and rounder.
10. Alcohol content: The alcohol content of the wine can contribute to its overall flavor, mouthfeel, and perceived quality.

Each of these features plays a role in determining the quality of a wine. For example, a wine with high levels of volatile acidity, total sulfur dioxide, or chlorides may be perceived as lower quality due to off-flavors, unpleasant aromas, or a harsh mouthfeel. Conversely, a wine with higher levels of citric acid or a lower pH may be perceived as higher quality due to its perceived freshness and acidity.    

## Q2. How did you handle missing data in the wine quality data set during the feature engineering process? Discuss the advantages and disadvantages of different imputation techniques.

Ans:
    
There are several techniques that can be used to handle missing data in the wine quality dataset, including:
1. Dropping missing values: This technique involves simply removing any rows or columns that contain missing data. The advantage of this technique is that it is simple and easy to implement. However, it can result in a loss of valuable data, especially if the number of missing values is high.
2. Mean imputation: This technique involves replacing missing values with the mean value of the non-missing data in the same feature. The advantage of this technique is that it is simple and can work well if the missing data is randomly distributed. However, it can lead to biased results if the missing data is not randomly distributed.
3. Median imputation: This technique involves replacing missing values with the median value of the non-missing data in the same feature. The advantage of this technique is that it is more robust to outliers than mean imputation. However, like mean imputation, it can lead to biased results if the missing data is not randomly distributed.
4. Mode imputation: This technique involves replacing missing values with the mode value (most frequent value) of the non-missing data in the same feature. The advantage of this technique is that it can work well for categorical data. However, like mean and median imputation, it can lead to biased results if the missing data is not randomly distributed.
5. K-nearest neighbor imputation: This technique involves finding the k-nearest neighbors of each missing value and using their values to impute the missing value. The advantage of this technique is that it can work well for non-linear relationships and can handle both numerical and categorical data. However, it can be computationally intensive and can result in biased results if the neighbors are not representative of the missing value.

In the wine quality dataset, we used mean imputation to handle missing data, as it is a simple and effective technique for handling missing numerical data. However, it is important to note that the choice of imputation technique will depend on the specific dataset and the characteristics of the missing data. It is important to carefully evaluate the advantages and disadvantages of each technique and select the one that is most appropriate for the given dataset.

```python
## function to plot QQ plots to check normality
import scipy.stats as stat
import pylab
import seaborn as sns
import matplotlib.pyplot as plt
def qq_plot(df, feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(df[feature],kde=True)
    plt.title(f'Histogram Plot for : {feature}')
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.title(f'Q-Q Plot for : {feature}')
    plt.show()
```

```python
## Iterate QQ plot on each of variable 
for col in df.columns:
    print(f'Feature : {col}')
    qq_plot(df,col)
    print('\n'+'='*100 + '\n')
```

## Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of features. What is the minimum number of principal components required to explain 90% of the variance in the data?

```python
# Ans:
    
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
# Load the wine quality dataset
wine_df = pd.read_csv('WineQT.csv')
# Separate the features from the target variable
X = wine_df.drop('quality', axis=1)
# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# Perform PCA to reduce the number of features
pca = PCA()
pca.fit(X_scaled)
# Calculate the cumulative explained variance ratio for each principal component
cumulative_variance_ratio = np.cumsum(pca.explained_variance_ratio_)
# Identify the minimum number of principal components required to explain 90% of the variance in the data
n_components = np.argmax(cumulative_variance_ratio >= 0.9) + 1
print(f"The minimum number of principal components required to explain 90% of the variance in the data is {n_components}")
```
