## Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.

Ans:
    
1. A Probability Mass Function (PMF) is a function that maps each possible value of a discrete random variable to its probability. It gives the probability that a discrete random variable X takes a particular value x, denoted by P(X=x). The PMF can be written as:

P(X=x) = f(x)

Where f(x) is the PMF of the random variable X.

For example, let's consider a six-sided fair die. The PMF of the random variable X, which represents the outcome of a single roll of the die, is given by:
P(X=1) = 1/6
P(X=2) = 1/6
P(X=3) = 1/6
P(X=4) = 1/6
P(X=5) = 1/6
P(X=6) = 1/6


2. A Probability Density Function (PDF) is a function that describes the probability distribution of a continuous random variable. It gives the relative likelihood that a continuous random variable X will take on a particular value x. Unlike the PMF, the PDF does not give the actual probability of a particular value, but instead gives the probability density, which can be interpreted as the likelihood of a value occurring in a small interval around x. The PDF can be written as:

f(x) = dF(x)/dx

Where f(x) is the PDF of the random variable X, and F(x) is the cumulative distribution function (CDF) of X.

For example, let's consider the height of adult males in a population. The PDF of the random variable X, which represents the height of a randomly selected adult male, might be a normal distribution with a mean of 5'10" (178 cm) and a standard deviation of 2.5 inches (6.35 cm). The PDF can be written as:

f(x) = (1/ (2πσ^2))^0.5 * e^(-(x-μ)^2 / (2σ^2))

where μ is the mean height and σ is the standard deviation of height in the population    

## Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

Ans:
    
The Cumulative Density Function (CDF) is a function that describes the cumulative probability distribution of a random variable X. It gives the probability that X takes a value less than or equal to a specified value x, denoted by F(x). The CDF can be written as:
F(x) = P(X ≤ x)

For a discrete random variable, the CDF is the sum of the PMF up to and including the value x. For a continuous random variable, the CDF is the integral of the PDF from negative infinity to x.

Here's an example of a CDF for a continuous random variable:
Suppose the weight of apples in a basket follows a normal distribution with a mean of 200 grams and a standard deviation of 25 grams. The CDF of this distribution can be calculated using the cumulative distribution function of the normal distribution.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
#parameters of the normal distribution
mu = 200
sigma = 25
# Define the CDF of the normal distribution
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 1000)
cdf = norm.cdf(x, mu, sigma)
# Plot the CDF
plt.plot(x, cdf)
plt.xlabel('Weight of Apples')
plt.ylabel('Cumulative Probability')
plt.title('CDF of the Weight of Apples')
plt.show()
```

Output:

The CDF shows that the probability of selecting an apple with weight less than or equal to 175 grams is around 13%, while the probability of selecting an apple with weight less than or equal to 225 grams is around 84%. The CDF can also be used to calculate probabilities of selecting apples with weights in a certain range or above a certain weight.

## Q7: Explain uniform Distribution with an example.

Ans:
    
The uniform distribution is a probability distribution that describes a situation where all values within a given range are equally likely to occur.

An example of a uniform distribution is rolling a fair six-sided die. Each of the six possible outcomes (1, 2, 3, 4, 5, and 6) has an equal chance of occurring, and therefore the distribution is uniform. The probability of rolling any one of the numbers is 1/6, or approximately 0.167.

In probability terms, if X is a continuous random variable with a uniform distribution over the interval [a, b], then the probability 
density function of X is given by:

f(x) = 1 / (b - a), for a ≤ x ≤ b
and f(x) = 0, for x < a or x > b

where a and b are the lower and upper bounds of the distribution.

```python
import numpy as np
# generate 10 random numbers from a uniform distribution between 0 and 1
uniform_data = np.random.uniform(0, 1, 10)
print(uniform_data)
```

This will output an array of 10 random numbers between 0 and 1. 
Note that the values are evenly distributed between 0 and 1, and each value has an equal probability of being generated.

## Q8: What is the z score? State the importance of the z score.

Ans:
    
The z-score, also known as the standard score, is a measure of how many standard deviations an individual observation or data point is from the mean of a population.

The formula for calculating the z-score is:

z = (x - μ) / σ

where:

x = the individual observation or data point

μ = the population mean

σ = the population standard deviation

The importance of the z-score lies in its ability to standardize data and allow for meaningful comparisons between observations from different populations or with different scales. By transforming raw scores into z-scores, we can compare observations that are measured on different scales or with different units of measurement. 
Additionally, the z-score allows us to determine the probability of an observation occurring in a particular distribution, as the z-score corresponds to a known area under the normal distribution curve. This can be useful in statistical hypothesis testing and in making inferences about population parameters from sample statistics.

## Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem

Ans:

CLT states that the sampling distribution of the mean of any independent, random variables will approximate a normal distribution, regardless of the shape of the underlying population distribution, provided that the sample size is large enough.

Significance:

1. It allows us to use the normal distribution to make inferences about population parameters, even when the population distribution is unknown or non-normal.

2. CLT forms the basis of many statistical techniques, such as hypothesis testing, confidence interval estimation, and regression analysis.

3. It also underlies many machine learning algorithms that rely on the assumption of normality, such as linear regression and logistic regression.

## Q10: State the assumptions of the Central Limit Theorem.

Ans:
    
The CLT relies on several assumptions, which include:
1. Independence: The observations or data points within each sample must be independent of each other. This means that the value of one observation should not influence the value of another observation within the same sample.
2. Sample Size: The sample size should be large enough to ensure that the sample mean is normally distributed, regardless of the shape of the population distribution. The rule of thumb is that the sample size should be greater than or equal to 30.
3. Finite variance: The population variance must be finite. This means that the population distribution should have a finite second moment, which is the square of the standard deviation.
4. Random Sampling: The samples should be drawn at random from the population, without bias or preference for any particular subset of the population.
